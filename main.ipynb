# Data Processing

```python
df = movie.merge(credits,on='title')
```

```python
df.describe()
```

```python
df.shape
```

```python
df = df[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True) #this is used to remove movies which have null values 
```

```python
import ast
```

```python
def convert(obj):
    l=[i['name'] for i in ast.literal_eval(obj) ]
    return l
```

```python
df['genres']=df['genres'].apply(convert)
```

```python
df.head()
```

```python
df['keywords']=df['keywords'].apply(convert)
```

```python
def convert3(obj):
    l=[]
    c=0
    for i in ast.literal_eval(obj):
        l.append(i['character'])
        c+=1
        if(c==3):
            break
    return l
```

```python
df['cast']=df['cast'].apply(convert3)
```

```python
def fatch_director(obj):
    l=[i['name'] for i in ast.literal_eval(obj) if i['department']=='Directing']
    return l
    
```

```python
df['crew']=df['crew'].apply(fatch_director)
```

```python
df['crew'][0]
```

```python
df['overview']=df['overview'].apply(lambda x:x.split())
```

```python
df.head()
```

```python
def rem_space(obj):
    return [i.replace(" ","") for i in obj]
```

```python
df['genres'] =df['genres'].apply(rem_space)
df['cast'] =df['cast'].apply(rem_space)
df['crew'] =df['crew'].apply(rem_space)
df['keywords'] =df['keywords'].apply(rem_space)
```

```python
df.head()
```

```python
df['tags']=df['overview']+df['genres']+df['keywords']+df['cast']+df['crew']
df.head()
```

```python
new_df=df[['movie_id','title','tags']] # creating a new dataframe which consits mov_id,title ana tags
```

```python
 new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x)) #making string from list  
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: x.lower()) #converting text of tags in lowercase to make easy for recomandation system
```

```python
new_df['tags'][0]
```

Given the possibility that in these tags, 'love,' 'loving,' and 'loved' may be treated as distinct words, it is necessary to equate them to their root term, 'love.'"

```python
import nltk
from nltk.stem.porter import PorterStemmer 
ps=PorterStemmer()
```

```python
def stem(text):
    y = [ps.stem(word) for word in text.split()]
    return " ".join(y)
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

# Vactorisation

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english',max_features=5000)
vector = tfidf.fit_transform(new_df['tags']).toarray()
```

```python
vector[0]
```

```python
vector.shape
```

```python
tfidf.get_feature_names_out() #this is used to get that 5000 words from vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity.shape
```

```python
similarity
```

# Main function

```python
list(enumerate(similarity[0]))
        #enumerate function gives tuple of index with its value 
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])[1:6]
    for i in distances:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Batman Begins")
```
